```python
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import ast
```

```python
obs = pd.read_excel("dataDU.xlsx", sheet_name="observed")
rand = pd.read_excel("dataDU.xlsx", sheet_name="random")
tags = pd.read_excel("dataDU.xlsx", sheet_name="tags")
```

```python
obs_clean = []

for i, j in enumerate(tags["tag"]):
    for k in tags["tag"][i+1:]:
        suma = (obs.loc[(obs['Tag1'] == j) & (obs["Tag2"] == k)]["duration_frames"].sum() + 
                obs.loc[(obs['Tag1'] == k) & (obs["Tag2"] == j)]["duration_frames"].sum())
        if suma > 0:
            obs_clean.append([j, k, suma])
            
obs_clean = pd.DataFrame(obs_clean, columns=["Tag1", "Tag2", "duration_frames"])
```

```python
obs_clean
```

```python
G = nx.from_pandas_edgelist(obs_clean, source='Tag1', target='Tag2',
                            edge_attr='duration_frames')

cols = []

for node in G:
    tag = tags.loc[tags["tag"] == node]["task_group"] 
    if (tag == "nurse").all():
        cols.append("lightblue")
    elif (tag == "forager").all():
        cols.append("orange")
    elif (tag == "queen").all():
        cols.append("green")
```

```python
widths = np.array([w for *_, w in G.edges.data('duration_frames')])

#pos = nx.shell_layout(G)
pos = nx.kamada_kawai_layout(G)

# nodes
nx.draw_networkx_nodes(G, pos, node_size=50, node_color=cols)

# edges
nx.draw_networkx_edges(G, pos, width=widths/1200) 

# labels
nx.draw_networkx_labels(G, pos, font_size=6, font_family="sans-serif")

#ax = plt.gca()
#ax.margins(0.08)
#plt.figure(figsize=(80, 60), dpi=500)
plt.axis("off")
#plt.tight_layout()

fig = plt.gcf()
fig.set_size_inches(32, 20)
fig.savefig('test2png.png', dpi=100)
```
